### 1.1
Considere a figura seguinte que representa uma sinusoide.

fig_trab

A partir da análise do gráfico determine os valores da amplitude ($A$), fase ($\phi$) e frequencia linear ($f$) associada à representação:

$ x(t) = A \cos(2\pi f t + \phi)$

Gere o código para implementar esta sinusóide.

```python
import numpy as np
import matplotlib.pyplot as plt

A= 4
Fs= 8000
T= 25
f= 1/T
deltaT = 10/8
phi= (deltaT*2*np.pi )/T
t= np.arange(-30, 40, 1/Fs)

## 2pi --- periodo
## phi --- delta t
## phi = 2pi*deltat/ periodo

x= A*np.sin(2*np.pi*f*t + phi)

plt.plot(t*1000,x)
plt.grid(True)
```

### 3. Série de Fourier

Considera o seguinte sinal:

$b(t) =  \sum_{k=-N}^{N}  \left(\frac{-1 }{j 2 k \pi}\right) e^{ j 2 \pi k f_0 t }$, onde $f_0=1$Hz. 

Gere o código para implementar b(t). Valide o resultado para $N \in \{1, 10, 1000 \}$.

```python
import numpy as np
import matplotlib.pyplot as plt
N = [1,10,1000]
f0= 1 #hz
fs = 2000 #hz

t= np.arange(0, 2, 1/fs)

for i in N:
    b=0
    for k in range(-i, i+1):
        if k != 0:
            b += (-1/(1j*2*k*np.pi)) * np.exp(1j*2*np.pi*k*f0*t)
    plt.subplot(3,1,N.index(i)+1)
    plt.plot(t,b)
```
